```python
pca = PCA()
```

```python
pca_x = pca.transform(tcounts)
```

```python
df_pcax = pd.DataFrame(pca_x)
```

```python
cov = pd.read_csv('cov.txt',sep = '\t', index_col = 0)
cov = cov.transpose()
cov = pd.DataFrame(cov)
```

```python
pr = pearsonr(df_pcax.iloc[:,0].values,cov.iloc[:,0].values)
```

```python
sig_cor = 0
for i in range(10):
    chk = 0
    for j in range(3):
        pr = pearsonr(df_pcax.iloc[:,i].values,cov.iloc[:,j].values)
        if chk == 0:
            if (abs(pr[0]) > 0.2) & (pr[1] < 0.05):
                sig_cor += 1
                print(str(i) + 'correlated with ' + str(j) + '\n')
                chk = 1
sig_cor        
```
